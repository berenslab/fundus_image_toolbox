```python
from fundus_image_toolbox.vessel_segmentation import load_segmentation_ensemble, ensemble_predict_segmentation, plot_masks, save_masks, load_masks_from_filenames
```

### Extract vessel mask from an image or image batch

Set `image` to an image path. 

Note: For the prediction, plotting and mask saving, `image` can as well be a list of paths, a list of image objects (PIL.Image, np.ndarray) or one image object. 

Define `device`, `SAVE_MASK` and `target_dir` as desired.

```python
image = "./imgs/drimdb_example.jpg"
```

```python
device = "cpu"
```

```python
SAVE_MASK = True
target_dir = "./masks"
```

Get the model and predict the masks for the images.

```python
ensemble = load_segmentation_ensemble(device=device)
```

```python
mask = ensemble_predict_segmentation(ensemble, image, device=device, threshold=0.5)
```

Plot the images and their corresponding masks.

```python
plot_masks(image, mask)
```

Save one mask to file

```python
if SAVE_MASK:
    save_masks(image, mask, target_dir)
```

Load a mask from file.

```python
mask = load_masks_from_filenames(image, target_dir)
```

```python
mask
```

```python
# Delete the saved masks again
import os

folder = target_dir
file = os.path.join(folder, "masks.pkl")

if os.path.exists(file):
    os.remove(file)
    os.rmdir(folder)
```
